```julia
for i in 1:size(d,1)
    net = FinNetValu.XOSModel(Mˢ, Mᵈ, Mᵉ, d[i,:], α1)
    f_i(x) = FinNetValu.valuation(net, x, a, α1) .- x
    Int =  (-3..3)
    r = roots(f_i, Int × Int × Int × Int × Int × Int, Newton, 1e-15)
    println("\nrun: ", i, " has ", length(r), " root(s):")
    for ri in r
        println(ri)
    end
    x1_1 = [el.lo + (el.hi-el.lo)/2 for el in r[1].interval];
    println("check (should be zero): ", round.(f_i(x1_1),digits=4))
end
```
